```python
#pone df1 y a su derecha df2, ->pone todos los datos(si hay datos nulos los pone igual) y respeta el orden de pers para acomodar los datos(aux:salta de categoria -1 a 2 )
s=pd.merge(pers, conicet,how="inner", on = "categoria_conicet_id") 
p3=pd.merge(pers, conicet,how="outer", on = "categoria_conicet_id")
p3.head(10)
```

```python
#pone df1 y a su derecha df2, ->pone unicamente las filas que coinciden (en las que no hay datos nulos) y respeta el orden de pers para acomodar los datos(aux:salta de categoria -1 a 2 )
s=pd.merge(pers, conicet,how="inner", on = "categoria_conicet_id") 
s.head(10)
```

pd.df.sort_values:
Ordena los valores del df-> orden se puede hacer tanto por columnas como por filas, ascendente o descendente, y eligiendo dónde deben ir los NaN
#axis: por defecto es 0. Con esto elegimos si se ordenan las columnas o las filas
#ascending: por defecto es True, por lo que se ordena de menor a mayor. Si es False se invierte el orden.
#na_position: por defecto es "last", por lo que todos los valores NaN se ordenan al final, pero se puede configurar como "first" para que aparezcan primero.

```python
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'col1': ['A', 'A', 'B', np.nan, 'D', 'C'],
    'col2': [2, 1, 9, 8, 7, 4],
    'col3': [0, 1, 9, 4, 2, 3],
    'col4': ['a', 'B', 'c', 'D', 'e', 'F']
})

print(df)
```

```python
# Ordenamos por columna 1:
#por defecto es ascending True, por eso en este caso aunq no puse nada ordena por letras del abecedario y ultimo los NaN
print(df.sort_values(by=['col1']))
```

```python
# Se pueden ordenar por más de una columna a la vez:
print(df.sort_values(by=['col1', 'col2'])) #si tengo 2 que me dan igual en la col1, los ordeno en base al ascendente de col2
```

```python
# Ordenamos de forma descendiente y con los NaN al principio:
print(df.sort_values(by='col1', ascending=False, na_position='first'))
#ordena en base a columna 1, de forma descendente #pone los valores nulos 1ros
```

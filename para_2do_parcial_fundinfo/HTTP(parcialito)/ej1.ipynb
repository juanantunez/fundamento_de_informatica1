```python
#b)
import requests #biblioteca que necesitamos para hacer pedidos
pedido = requests.get("https://ponyweb.ml/v1/character/all")#“get” es pedir un recuso() #ingreso URL de dato->que en este caso pide una prenda en especifico
print(pedido.status_code) # pide el solo el nro del status_code q devolvia entre <Response[nro]>
print(pedido.headers["Content-Type"])#Las cabeceras son todas las cosas que tira el diccionario al hacer el atributo .headers. Estasson:X-Powered-By, Content-Length,  Date, Content-Type. entre corchetes pongo Content-Type que es el que quiero que me devuelva
```

```python
#e)

ponies=requests.get("https://ponyweb.ml/v1/character/all")

#todo lo que describo ahora es para ponerle el nombre al archivo:
#"data" es la clave de la lista de diccionarios de todos los ponies, 
# tomo la posicion "[43]" para obtener al pony 44(dado que posiciones cuentan posicion 0)
#toma dentro del diccio correspondiente al pony 44, la clave "["name"]->esto devuelve el nombre de dicho pony 
#le agrego extension ".txt"

#todo esto que puse hasta recien va a ser un archivo al que le voy a escribir "w" ciertas cosas

#a este archivo le voy a escribir "write" toda la data del pony 44


with open(ponies.json()["data"][43]["name"]+".txt", "w") as archivo:
    archivo.write(str(ponies.json()["data"][43]))
```
